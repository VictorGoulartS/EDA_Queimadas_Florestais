## 1. Introdução
As florestas brasileiras sofrem com a falta de conscientização da sociedade, embora ocorra incêndios provocado pela natureza de forma regular, está havendo uma contribuição humana mais influente, devastando os ecossistemas do país. A grave situação de queimadas junto com o desmatamento insustentável estão resultando na morte de um grande número de plantas e animais.

Neste sentido, o projeto tem o objetivo de realizar uma análise exploratória (EDA - Exploratory Data Analysis) acerca dos dados das queimadas florestais no Brasil, com o foco de entender melhor onde e quando ocorrem as queimadas florestais, para que assim, seja mais fácil de realizar medidas para preveni-las.

**Fonte de dados**

Os dados aplicados nesse projeto foram adquiridos no site oficial do governo brasileiro, onde apresenta o número de queimadas florestais que ocorreram em cada estado do Brasil. Estes dados se referem às queimadas florestais que ocorreram entre 1998 até 2017.

**Escopo do projeto**

Com o uso dos dados disponibilizados pelo governo brasileiro, será analisado quais as regiões que são mais afetadas pelas queimadas e em quais épocas do ano as queimadas são mais frequentes. 

**Método de análise**

O método de análise utilizado no projeto é a análise descritiva pois, a análise descritiva visa descrever as principais tendências nos dados e em observar situações que levam a novos fatos. Considerando isso, ao utilizarmos essa análise, vamos chegar no nosso objetivo de entender melhor como ocorrem as queimadas nas florestas do Brasil. 

**Contribuições**

A análise de dados no nosso projeto irá gerar insights nos quais poderemos identificar as áreas com maior e menor quantidade de queimadas florestais reportadas, para que assim, seja mais fácil de saber onde estão ocorrendo essas queimadas e quais áreas precisam que aumentem mais a quantidade de pessoas reportando as queimadas, caso contrário, não será possível prevenir que isso aconteça.

### 4.5 Alterações Dados

```python
# Vamos traduzir o nome das colunas para melhor entendimento dos dados.
dados = dados.rename(columns={'year': 'ano',
                              'state': 'estado',
                              'month': 'mes',
                              'number': 'quantidade_queimadas'})
dados.head()
```

### 4.6. Adição de Dados

```python
# Exibe todos os estados cadastrados no dataset, todos os 26 estados do Brasil e o Distrito Federal.
dados['estado'].unique()
```

```python
# Preenche as regiões de cada estado.
estados_regioes = {'Acre': 'Norte', 'Amapa': 'Norte', 'Amazonas': 'Norte', 'Pará': 'Norte',
                   'Rondonia': 'Norte', 'Roraima': 'Norte', 'Tocantins': 'Norte', 
                   'Alagoas': 'Nordeste', 'Bahia': 'Nordeste', 'Ceara': 'Nordeste', 'Maranhao': 'Nordeste', 
                   'Paraiba': 'Nordeste', 'Pernambuco': 'Nordeste', 'Piauí': 'Nordeste', 
                   'Rio Grande do Norte': 'Nordeste', 'Sergipe': 'Nordeste', 
                   'Distrito Federal': 'Centro-Oeste', 'Goias': 'Centro-Oeste', 
                   'Mato Grosso': 'Centro-Oeste', 'Mato Grosso do Sul': 'Centro-Oeste',
                   'Espirito Santo': 'Sudeste', 'Minas Gerais': 'Sudeste', 'Rio de Janeiro': 'Sudeste', 'Sao Paulo' : 'Sudeste', 
                   'Paraná': 'Sul', 'Rio Grande do Sul': 'Sul', 'Santa Catarina': 'Sul'}

# Cria coluna de acordo com os dados preenchidos acima. 
dados['regiao_estado'] = dados['estado'].map(estados_regioes)
print(dados)
```

## 4.8. Visualização dos Dados

Segue abaixo, a visualização dos dados visando responder as perguntas realizadas no começo do projeto.

## 4.8.1. Quais os meses com maior incidência de queimadas reportadas?

```python
queimadas_mes = dados.groupby(['mes']).sum()['quantidade_queimadas']

#Ordena pela ordem dos meses.
meses = dados['mes'].unique()

queimadas_mes = queimadas_mes.reindex(meses)
queimadas_mes
```

```python
plt.rc('figure', figsize=(15,6))
plt.bar(meses, queimadas_mes, color = '#b91010')
plt.xticks(rotation=30, horizontalalignment="center")
plt.title('Total de Queimadas nas Florestas reportadas no Brasil entre 1998 à 2017 por mês')
plt.xlabel('Meses')
plt.ylabel('Quantidade de queimadas')
plt.show()
```

Podemos verificar que os meses de Fevereiro, Junho e Novembro são os meses que tem as maiores quantidades de queimadas florestais, eles possuem dados muito próximos, mas o mês de Junho é o mês que mais ocorreu queimadas reportadas durante esses anos.

## 4.8.2. Quais os estados com maior incidência de queimadas reportadas?

```python
queimadas_estado = dados.groupby(['estado']).sum()['quantidade_queimadas']

# Ordena os estados por maior quantidade de queimadas.
queimadas_estado = queimadas_estado.sort_values(ascending=False)
queimadas_estado
```

```python
plt.rc('figure', figsize=(15,6))
plt.xticks(rotation=90, horizontalalignment="center")
plt.bar(queimadas_estado.index, queimadas_estado, color = '#b91010')
plt.title('Total de Queimadas nas Florestas reportadas no Brasil entre 1998 à 2017 por estado')
plt.xlabel('Estados')
plt.ylabel('Quantidade de queimadas')
plt.show()
```

Concluímos que dois dos três estados da região do Centro-Oeste estão entre os 3 estados que mais possuem queimadas florestais reportadas e que o estado de São Paulo ficou em primeiro lugar neste quesito.

## 4.8.3. Quais as regiões com maior incidência de queimadas reportadas?

```python
queimadas_regiao = dados.groupby(['regiao_estado']).sum()['quantidade_queimadas']

regioes = dados['regiao_estado'].unique()

queimadas_regiao
```

### Horizontal Bar Chart

```python
plt.rc('figure', figsize=(15,6))
plt.barh(regioes, queimadas_regiao, color = '#b91010')
plt.xticks(horizontalalignment="center")
plt.title('Total de Queimadas nas Florestas reportadas no Brasil entre 1998 à 2017 por região')
plt.xlabel('Quantidade de queimadas')
plt.ylabel('Regiões')
plt.show()
```

Para que possamos entender melhor a proporção de uma região para outra, segue abaixo, um Tree Map sobre os dados evidenciados acima:

### Tree Map

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify
import seaborn as sb

regioes = dados['regiao_estado'].unique()

queimadas_regiao = dados.groupby(['regiao_estado']).sum()['quantidade_queimadas']

# Criar treemap da distribuição por categoria
plt.rc('figure', figsize=(8,6))
squarify.plot(sizes=queimadas_regiao, label=queimadas_regiao.index, alpha=0.8, color=sb.color_palette("tab10",  len(queimadas_regiao)))
plt.axis('off')
plt.title('Distribuição de Quantidade de Queimadas Reportadas por Região no Treemap')
plt.show()
```

Podemos identificar que mesmo dois dos três estados da região do Centro-Oeste estarem entre os 3 estados que mais tiveram queimadas florestais reportadas, ele ficou em 3º lugar, atrás do Nordeste, em primeiro e do Norte, em segundo.

## 4.8.4. Qual a incidência de queimadas reportadas entre o período de 1998 à 2017?

```python
queimadas_ano = dados.groupby(['ano']).sum()['quantidade_queimadas']
queimadas_ano.index
```

```python
queimadas_ano_label = [str(x) for x in queimadas_ano.index]

plt.rc('figure', figsize=(15,6))
plt.plot(queimadas_ano_label, queimadas_ano, color = '#b91010')
plt.title('Variação da quantidade de queimadas das florestas no Brasil ao passar dos anos')
plt.xlabel('Anos')
plt.ylabel('Quantidade de queimadas')
plt.grid(True)
plt.show()
```

Podemos identificar que o ano de 2003 teve um pico de queimadas florestais reportadas no Brasil. Também é possível identificar que entre o ano de 2002 à 2017, o ano de 2008 foi o ano que teve o menor número de queimadas florestais reportadas.
